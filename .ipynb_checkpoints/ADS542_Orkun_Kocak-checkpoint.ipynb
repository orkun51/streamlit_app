## Data Cleaning <a class="anchor" id="2">
<p>Load the dataset and check for any missing values or inconsistencies.</p>

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Separate numeric and non-numeric columns
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
non_numeric_cols = data.select_dtypes(exclude=['int64', 'float64']).columns

# Handle negative values in numeric columns
negative_value_counts = (data[numeric_cols] < 0).sum()
negative_cols = negative_value_counts[negative_value_counts > 0].index.tolist()
for col in negative_cols:
    min_val = data[col].min()
    data[col] = data[col] + abs(min_val) + 1

# Verify all values are now non-negative
assert (data[negative_cols] >= 0).all().all(), "There are still negative values in the dataset"
```

##  Feature Selection <a class="anchor" id="2">
<p>Feature selection techniques to select the most relevant features.</p>

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split
# Feature selection
X = data.drop('y', axis=1)
y = data['y']
selector = SelectKBest(chi2, k=10)
X_new = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support(indices=True)]
print("Selected features:", selected_features)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.3, random_state=42)
```

##  Model Selection <a class="anchor" id="3">
<p>Proceed with model selection using the selected and scaled features.</p>

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from scipy.stats import randint, uniform
import pickle

# Define the models
models = {
    'Random Forest': RandomForestClassifier(),
    'Neural Network': MLPClassifier(max_iter=1000),
    'Gradient Boosting': GradientBoostingClassifier()
}

# Define the parameter distributions for RandomizedSearchCV
param_distributions = {
    'Random Forest': {
        'n_estimators': randint(50, 200),
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': randint(2, 11),
        'min_samples_leaf': randint(1, 5)
    },
    'Neural Network': {
        'hidden_layer_sizes': [(50,), (100,), (50, 50)],
        'activation': ['relu', 'tanh'],
        'solver': ['adam', 'sgd'],
        'alpha': uniform(0.0001, 0.001)
    },
    'Gradient Boosting': {
        'n_estimators': randint(50, 200),
        'learning_rate': uniform(0.01, 0.2),
        'max_depth': randint(3, 8)
    }
}

# Train and tune the models using RandomizedSearchCV
best_models = {}
for model_name, model in models.items():
    randomized_search = RandomizedSearchCV(model, param_distributions[model_name], n_iter=50, cv=5, scoring='f1', random_state=42, n_jobs=-1)
    randomized_search.fit(X_train, y_train)
    best_models[model_name] = randomized_search.best_estimator_
    print(f"Best parameters for {model_name}: {randomized_search.best_params_}")

# Evaluate the models
results = {}
for model_name, model in best_models.items():
    y_pred = model.predict(X_test)
    results[model_name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'Classification Report': classification_report(y_test, y_pred, target_names=['No', 'Yes']),
        'Confusion Matrix': confusion_matrix(y_test, y_pred)
    }
    print(f"\nClassification report for {model_name}:\n{results[model_name]['Classification Report']}")
    print(f"Confusion matrix:\n{results[model_name]['Confusion Matrix']}")

# Save the best model to a file
best_model_name = max(results, key=lambda x: results[x]['F1 Score'])
with open('best_model.pkl', 'wb') as model_file:
    pickle.dump(best_models[best_model_name], model_file)
print(f"Best model saved: {best_model_name}")
```

Creating Pipeline

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler

# Create a pipeline with SMOTE and the model
def create_pipeline(model):
    pipeline = Pipeline([
        ('smote', SMOTE(random_state=42)),
        ('scaler', StandardScaler()),
        ('classifier', model)
    ])
    return pipeline

# Example usage of the pipeline with one of the best models
best_model = best_models[best_model_name]
pipeline = create_pipeline(best_model)
pipeline.fit(X_train, y_train)
```
